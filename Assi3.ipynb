```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Libraries imported.")
```

```python
#Load Dataset
df = pd.read_csv("/content/student_performance.csv")
print(df.head())
```

```python
# Features and target
X = df[['math', 'science', 'english', 'attendance']]
y = df['result']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```python
# KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print("Model trained.")
```

```python
# Prediction
y_pred = knn.predict(X_test)
print("Predictions:", y_pred[:10])
```

```python
# Show Correct and Wrong Predictions
results = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})

results['Status'] = results.apply(
    lambda r: 'Correct' if r.Actual == r.Predicted else 'Wrong',
    axis=1
)

print(results)
```

```python
#Accuracy and Evaluation
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

Conclusion:
The KNN algorithm was successfully used to predict student performance pass or fail based on marks and attendance.
The model achieved good accuracy, showing that KNN works well for small, simple datasets.
This approach is easy to understand and suitable for beginner-level machine learning projects.
