```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
print('Libraries imported.')
```

```python
url = '/content/test.csv'
columns = ['sepal_length','sepal_width','petal_length','petal_width','class']
df = pd.read_csv(url, names=columns)
df.head()
```

```python
print('Shape:', df.shape)
print('\nColumn Names:', df.columns.tolist())
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include=['float64','int64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill categorical values with mode
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

df.isnull().sum()
```

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print('Duplicates removed:', before - after)
```

```python
encoder = LabelEncoder()
df['class'] = encoder.fit_transform(df['class'])
df.head()
```

```python
df.to_csv('assignment2_cleaned', index=False)
print('Saved: assignment2_cleaned_iris.csv')
```

Conclusion : This code performs essential data preprocessing to prepare the dataset for machine learning. It loads the CSV file, explores the data structure, and checks for missing values. Missing numerical values are filled with the mean, while categorical values are filled using the mode. Duplicate records are removed to improve data quality. The class column is encoded into numerical form using Label Encoding. Finally, the cleaned dataset is saved for further analysis or model training.
